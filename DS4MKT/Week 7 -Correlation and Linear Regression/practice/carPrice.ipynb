```python
import warnings
warnings.filterwarnings('ignore')

#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('CarPrice_Assignment.csv')

cars.head()
```

```python
cars.shape
```

```python
cars.describe()
```

```python
cars.info()
```

```python
CompanyName = cars['CarName'].apply(lambda x : x.split(' ')[0])
cars.insert(3,'CompanyName', CompanyName)
cars.drop(['CarName'], axis=1, inplace=True)
cars.head()
```

```python
#See all unique values for company name cars
cars.CompanyName.unique()
```

Agora vamos arrumar alguns nomes das companias

```python
cars.CompanyName = cars.CompanyName.str.lower()

#Criar uma função para trocar cada nome da tabela
def replace_name(a,b):
    cars.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

cars.CompanyName.unique()
```

```python
plt.figure(figsize=(14,7))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(cars.price)

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=cars.price)

plt.show()
```

```python
print(cars.price.describe(percentiles = [0.25,0.50,0.75,0.85,0.90,0.95]))
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
plt.title('Companies')
sns.countplot(cars.CompanyName)

plt.subplot(1,3,2)
plt.title('Fuel Type')
sns.countplot(cars.fueltype)

plt.subplot(1,3,3)
plt.title('Body Type')
sns.countplot(cars.carbody)

plt.show()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('Symboling Histogram')
sns.countplot(cars.symboling, palette=("cubehelix"))

plt.subplot(1,2,2)
plt.title('Symboling vs Price')
sns.boxplot(x=cars.symboling, y=cars.price, palette=("cubehelix"))

plt.show()
```
